```python
import json
data = json.load(open('data/processed_spice_data/dev_each_type_50.json', 'r'))
```

```python
conversations = []

last_question = ''
last_ID = ''
conversation = None

for qa in data:
    if not last_question:
        conversation = [qa]
        qa['last_ID'] = ''
    elif not qa['question'].startswith(last_question):
        conversations.append(conversation)
        conversation = [qa]
        qa['last_ID'] = ''
    else:
        conversation.append(qa)
        qa['last_ID'] = last_ID
    last_ID = qa['turnID']
    last_question = qa['question'][:-6]

# for i in range(10):
#     for qa in conversations[i]:
#         print(qa['question'])
#     print()

conversations[:10]
```

```python
batches = []
batch = [(i, 0) for i in range(20)]
pointer = 20
while batch:
    batches.append([conversations[i][j] for i, j in batch])
    for _ in range(len(batch)):
        i, j = batch.pop(0)
        if len(conversations[i]) > j + 1:
            batch.append((i, j + 1))
        else:
            if pointer < len(conversations):
                batch.append((pointer, 0))
                pointer += 1
```

```python
json.dump(batches, open('data/processed_spice_data/batches.json', 'w'), indent=2)
```
